## Mask Filling

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("The climate in Peru is normally <mask>.", top_k=2) #Display two possible answers for the masked token
```

## Named entity recognition (ner)

```python
from transformers import pipeline

ner = pipeline( "token-classification", model="dslim/bert-base-NER", aggregation_strategy="simple")
result = ner("My pet is called Fredy and he is a golden retriever.")
print(result)
```

## Question answering

```python
from transformers import pipeline

generator = pipeline( "text-generation", model="gpt2")
prompt = """
Question: What is the capital of France?
Context: Paris is the capital and most populous city of France.
Answer:
"""

result = generator(prompt, max_new_tokens=20)

print(result[0]["generated_text"])
```

## Summarization

```python
from transformers import pipeline

generator = pipeline( "text-generation",model="HuggingFaceTB/SmolLM2-360M")

text = """
Peru has experienced significant economic and social transformation during recent years. The country has become one of the fastest-growing economies in South America, driven by industries such as mining, agriculture, tourism, and fisheries.
Engineering plays a crucial role in Peru’s future development. Civil, environmental, mining, and agricultural engineering are especially important because of the country’s diverse geography and natural resources.
Like many developing countries, Peru must balance technological advancement with the preservation of its environment and cultural heritage.
"""

prompt = f"""
Summarize the following text in 3 sentences:

{text}

Summary:
"""
result = generator( prompt, max_new_tokens=80, do_sample=False)
print(result[0]["generated_text"])
```

## Translation

```python
from transformers import pipeline

generator = pipeline("text-generation", model="HuggingFaceTB/SmolLM2-360M")

prompt = "Translate to English: Mi gato es de color negro. Answer:"

result = generator(prompt, max_new_tokens=20, do_sample=False)

print(result[0]["generated_text"])
```

## Image classification

```python
from transformers import pipeline

image_classifier = pipeline(
    task="image-classification", model="google/vit-base-patch16-224"
)
result = image_classifier(
    "https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/pipeline-cat-chonk.jpeg"
)
print(result)
```

## Automatic speech recognition

```python
from transformers import pipeline

transcriber = pipeline(
    task="automatic-speech-recognition", model="openai/whisper-large-v3"
)
result = transcriber(
    "https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/mlk.flac"
)
print(result)
```
